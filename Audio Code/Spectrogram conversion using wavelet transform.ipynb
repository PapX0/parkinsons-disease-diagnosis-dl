```python
import os
import pandas as pd
import numpy as np
from scipy.signal import cwt, morlet
import matplotlib.pyplot as plt

def generate_combined_spectrogram(audio_data, file_name, scales, output_directory):
    try:
        # Apply Continuous Wavelet Transform with Morlet wavelet
        spectrogram = cwt(audio_data, morlet, scales)

        # Display the spectrogram with logarithmic scale
        plt.imshow(np.log1p(np.abs(spectrogram)), cmap='viridis', aspect='auto',
                   extent=[0, len(audio_data), min(scales), max(scales)])
        plt.colorbar(label='Log Magnitude')
        plt.title(f'Spectrogram of {file_name}')
        plt.xlabel('Time')
        plt.ylabel('Scale')

        # Save the spectrogram image
        output_filename = f"{file_name}_spectrogram.png"
        output_path = os.path.join(output_directory, output_filename)
        plt.savefig(output_path)
        plt.close()  # Close the plot to prevent displaying it

        print(f'Spectrogram saved as {output_path}')
    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# Paths
csv_file_path = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\Testing\testing.csv"
output_directory = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\Testing"

# Load CSV file
dataset = pd.read_csv(csv_file_path)

# Set up the scales for CWT
scales = np.arange(1, 22)

# Iterate over each row in the dataset
for index, row in dataset.iterrows():
    audio_data = row.values[2:]  # Extract audio data from the row, skipping first two columns
    generate_combined_spectrogram(audio_data, f"Row_{index}", scales, output_directory)
```

Shearlet


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sporco import plot
from sporco import util
from sporco.admm import cbpdn

def generate_combined_spectrogram(audio_data, file_name, output_directory):
    try:
        # Set up Shearlet Transform parameters
        N = audio_data.shape[0]
        M = 1
        opt = cbpdn.ConvBPDN.Options({'Verbose': False, 'MaxMainIter': 200,
                                      'RelStopTol': 5e-3, 'AuxVarObj': False})

        # Construct dictionary
        D = util.convdicts.Shearlet((N, M), scales)

        # Solve Convolutional BPDN problem
        b = np.expand_dims(audio_data, axis=1)
        X = cbpdn.cbpdn(D, b, 10.0, opt)

        # Display the coefficients
        plot.imview(np.log1p(np.abs(X.squeeze())), title='Shearlet Coefficients',
                    cmap='viridis', axis='tight')

        # Save the spectrogram image
        output_filename = f"{file_name}_shearlet_coefficients.png"
        output_path = os.path.join(output_directory, output_filename)
        plt.savefig(output_path)
        plt.close()  # Close the plot to prevent displaying it

        print(f'Shearlet coefficients saved as {output_path}')
    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# Paths
csv_file_path = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\testing.csv"
output_directory = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\Shearlet transform spctrgrm\Testing spectrogram"

# Load CSV file
dataset = pd.read_csv(csv_file_path)

# Set up the scales for Shearlet Transform
scales = [4, 4]

# Iterate over each row in the dataset
for index, row in dataset.iterrows():
    audio_data = row.values[2:]  # Extract audio data from the row, skipping first two columns
    generate_combined_spectrogram(audio_data, f"Row_{index}", output_directory)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import contourlet_transform

def generate_contourlet_transform(audio_data, file_name, output_directory):
    try:
        # Convert audio data to image (reshape as needed)
        audio_img = np.reshape(audio_data, (1, len(audio_data)))

        # Perform Contourlet Transform (Ridgelet Transform in scikit-image)
        contourlet_img = contourlet_transform(audio_img)

        # Plot and save the Contourlet Transform image
        plt.imshow(np.abs(contourlet_img), cmap='viridis')
        plt.colorbar(label='Magnitude')
        plt.title(f'Contourlet Transform of {file_name}')
        plt.savefig(os.path.join(output_directory, f"{file_name}_contourlet_transform.png"))
        plt.close()

        print(f'Contourlet Transform image saved for {file_name}')
    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# File paths
csv_file_path = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\testing.csv"
output_directory = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted\Contourlet transform spectrogram"

# Load CSV file
dataset = pd.read_csv(csv_file_path)

# Iterate over each row in the dataset
for index, row in dataset.iterrows():
    audio_data = row.values[2:]  # Extract audio data from the row, skipping first two columns
    generate_contourlet_transform(audio_data, f"Row_{index}", output_directory)
```
