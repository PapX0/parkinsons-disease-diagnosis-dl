```python
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler

# CSV file path
csv_file_path = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\SAKAR pd_speech_features.csv"

# Output directories
output_dir_0 = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted"
output_dir_1 = r"C:\Users\Lenovo\Downloads\PD AUDIO DATA\Sakar pd speech time series converted"

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Drop the first two columns
df = df.iloc[:, 2:]

# Normalize features
scaler = MinMaxScaler()
df[df.columns[:-1]] = scaler.fit_transform(df[df.columns[:-1]])

# Create output directories if they don't exist
os.makedirs(output_dir_0, exist_ok=True)
os.makedirs(output_dir_1, exist_ok=True)

# Split the DataFrame based on the value of "class"
class_0_df = df[df['class'] == 0].drop(columns=['class'])
class_1_df = df[df['class'] == 1].drop(columns=['class'])

# Write the split DataFrames to separate CSV files
class_0_df.to_csv(os.path.join(output_dir_0, 'class_0.csv'), index=False)
class_1_df.to_csv(os.path.join(output_dir_1, 'class_1.csv'), index=False)
```
